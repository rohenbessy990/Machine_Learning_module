```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
day=[1,2,3,4,5,6,7]
temp=[50,51,52,48,47,49,46]
```

```python
plt.plot(day,temp)
plt.show()
```

```python
plt.plot(day,temp,color='red',marker='o', linestyle='solid', linewidth=2,) #linestyle : dotted,solid ,dash dotted
plt.show()
```

```python
plt.xlabel('day',fontsize=20)
plt.ylabel('temp',fontsize=20)
plt.title("weather info")
plt.plot(day,temp,color='Red',marker='o', linestyle='solid', linewidth=2,alpha=0.6) #Linestyle:dotted,solid ,dash dotted
plt.show()  #Alpha : color intensity 0 to 1 value
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[20,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('days')
plt.ylabel('Temperature')
plt.title("Weather Info")
plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")
plt.legend(loc='best',shadow=True,fontsize='large')
plt.grid()
plt.show()
```

```python
 # Simple bar chart showing revenues of major US tech companies

company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,25,34,12]
loss=[23,12,24,8]
```

```python
plt.xlabel("Major Companies")
plt.ylabel("Profit and Revenue")
plt.title("Company Revenue 2022")
plt.bar(company,revenue,label="Revenue")
plt.bar(company,profit,label="Profit")
plt.bar(company,loss,label="loss")
plt.legend()
plt.show()
```

```python
#have to create an array to seperate both revenue and profit to show in same bar graph
xpos=np.arange(4)
```

```python
xpos=np.arange(len(company))
```

```python
xpos
```

```python
plt.xticks(xpos,company)
plt.xlabel("Company Name")
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="profit")
plt.legend()
```

```python
plt.xticks(xpos,company)
plt.xlabel("Major Companies")
plt.ylabel("Profit and Revenue")
plt.title("Company Revenue 2022")
plt.bar(xpos-0.2,revenue,width=0.2,label="Revenue")
plt.bar(xpos+0,profit,width=0.2,label="Profit")
plt.bar(xpos+0.2,loss,width=0.2,label="loss")
plt.legend()
plt.show()
```

```python
#To plot bar horizontally givr barh instaed of bar
plt.barh(company,revenue)
```

```python
#histogram
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar)
```

```python
plt.hist(blood_sugar,rwidth=0.8,color="red")
```

```python
plt.hist(blood_sugar,rwidth=0.8,bins=3,color="black")
```

```python
plt.hist(blood_sugar,bins=[80,100,125,150] ,rwidth=0.8) # by default number of bins is set to 10
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.8,label=['men','women'],color=["black","grey"])
plt.legend()
plt.show()
```

```python
# Draw pie chart to track down home expenses
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.figure(figsize=(10,7))
plt.axis('equal')
plt.pie(exp_vals,labels=(exp_labels),radius=1,autopct="%0.2f%%",shadow=True,explode=[0.1,0,0,0.22,0],startangle=90)  #autopct is for giving percentage inside piechart
plt.legend(shadow=True)#Explode to seperate each thing
#plt.show()#startangle to make certain angles
plt.savefig("piechart_new.jpg")# if we give plt.show the file image cant be seen in saved file
```
